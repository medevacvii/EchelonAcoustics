```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

DATA_DIR = "../data/frog_clips"

classes = ["CONTROL", "TOAD-WEST"]

print("Available classes:", classes)

for cls in classes:
    folder = os.path.join(DATA_DIR, cls)
    files = os.listdir(folder)
    print(f"\nClass: {cls}")
    print(f"Number of audio files: {len(files)}")
    print(files[:5])
```

## Visualization sample audio

```python
sample_path = "../data/frog_clips/TOAD-WEST/" + os.listdir("../data/frog_clips/TOAD-WEST")[0]
y, sr = librosa.load(sample_path, sr=22050)

plt.figure(figsize=(12,3))
librosa.display.waveshow(y, sr=sr)
plt.title("Sample Waveform")
plt.show()

plt.figure(figsize=(12,4))
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
S_dB = librosa.power_to_db(S, ref=np.max)
librosa.display.specshow(S_dB, sr=sr, x_axis="time", y_axis="mel")
plt.title("Mel Spectrogram")
plt.colorbar()
plt.show()
```
